```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv('./Data/Culinária Peruana.csv')
df = df.drop(columns=['Carimbo de data/hora', 'Unnamed: 12', 'Unnamed: 13'])
print(f'A pesquisa possui {df.shape[1]} perguntas e foi respondido por {df.shape[0]} pessoas.')
print('confira abaixo os resultados: ')
df
```

```python
quantidade_idade = df['Qual sua idade?'].value_counts()
print("Distribuição de Idades dos Respondentes:\n")
print(f"Total de Respostas: {quantidade_idade.sum()}")
faixas_etarias = ['18 a 30', '31 a 40', '41 a 50', '51 a 60', '61 ou mais']
for faixa in faixas_etarias:
    count = quantidade_idade.get(faixa, 0)  # Usando get para evitar KeyError se a faixa não existir
    print(f"Faixa etária '{faixa}': {count} ({count/quantidade_idade.sum()*100:.1f}%)")

fig, ax = plt.subplots(figsize=(10, 10))
sns.set(style="whitegrid")
ax.pie(quantidade_idade, labels=quantidade_idade.index, autopct='%1.1f%%', startangle=90)
ax.set_title('Distribuição por Idade')

plt.savefig('./Gráficos/Distribuição por Idade.png')
plt.show()
```

### Insights:
70% dos participantes da pesquisa possuem idade entre 18 a 30 anos.

```python
quantidade_genero = df['Qual seu gênero?'].value_counts()
print("Distribuição de Gênero dos Respondentes:\n")
print(f"Total de Respostas: {quantidade_genero.sum()}")
for genero, count in quantidade_genero.items():
    print(f"Gênero '{genero}': {count} ({count/quantidade_genero.sum()*100:.1f}%)")
    
plt.figure(figsize=(10, 10))
ax = df['Qual seu gênero?'].value_counts().plot(kind='bar', title='Distribuição de Gênero')
ax.set_xticklabels(ax.get_xticklabels(), rotation=360)
ax.set_xlabel('')
ax.set_title('Distribuição de Gênero')

plt.savefig('./Gráficos/Distribuição por Gênero.png')
plt.show()
```

### Insights:
Cerca de 70.25% dos respondentes são do sexo feminio.

```python
contagem_exp = df['Você já experimentou alguma comida peruana?'].value_counts()
print('Distribuição dos respondentes sobre se já experimentaram alguma comida peruana:')
print(f"Total de Respostas: {contagem_exp.sum()}")
for resposta, count in contagem_exp.items():
    print(f"Respostas '{resposta}': {count} ({count/contagem_exp.sum()*100:.1f}%)")

fig, ax = plt.subplots(figsize=(10, 10))

ax.pie(contagem_exp, labels=contagem_exp.index, autopct='%1.1f%%', startangle=90)
ax.set_title('Já experimentou alguma comida peruana?')

plt.savefig('./Gráficos/Experiência com comida peruana.png')
plt.show()
```

### Insight:
- Aproximadamente 42.1% dos respondentes já experimentaram comida peruana, o que indica uma boa exposição a essa cultura culinária entre o público pesquisado. Isso pode refletir uma tendência de abertura a experiências gastronômicas internacionais.
 - Em paralelo, há um segmento considerável (38.8%) que ainda ``não`` experimentou comida peruana, Estratégias de educação sobre a culinária peruana poderiam potencialmente converter parte desse grupo em novos clientes ou entusiastas da culinária.
 - Os 19.0% que responderam ``Não sei dizer`` podem indicar falta de conhecimento sobre o que constitui a comida peruana ou incerteza sobre se o que comeram era autenticamente peruano. Isso sugere que pode haver uma necessidade de mais informação e educação sobre essa culinária entre o público em geral.

```python
esc_comida = df['Quais dessas opções você acha que são comidas clássicas da culinária peruanas ?'].value_counts()
print(esc_comida)
print("\n")
print(f"Total de Respostas: {esc_comida.sum()}")
print('Número e pocentagem de menções das comidas clássicas da culinária peruana:')
comidas_frequentes = ['Chicha morada', 'Ceviche', 'Tamal', 'Lomo saltado']
for comida in comidas_frequentes:
    total_mencoes = esc_comida[esc_comida.index.str.contains(comida)].sum()
    print(f"Menções de '{comida}': {total_mencoes} ({total_mencoes/esc_comida.sum()*100:.1f}%)")

plt.figure(figsize=(10, 8))
esc_comida.sort_values().plot(kind='barh')
plt.title('Respostas sobre comidas clássicas da culinária peruana')
plt.ylabel('Comida mencionada')
plt.xlabel('Número de votos')

plt.savefig('./Gráficos/Comidas clássicas da culinária peruanas.png')
plt.show()
```

### Insights:
De acordo com a responsável do artigo, as respostas corretas seriam ``Chicha morada``, ``Ceviche``, ``Tamal`` e ``Lomo saltado``. Contudo, nas 121 amostras, não houve essa combinação específica. Ainda assim, podemos extrair alguns insights:
 - ``Chicha morada`` e ``Ceviche`` são claramente mais reconhecidos ou associados à culinária peruana pelos respondentes.
 - Por outro lado,``Lomo saltado`` foi citado por apenas 3 participantes da pesquisa, o que corresponde a pouco mais de 2% de todos os respondentes.

```python
coluna = 'Você acha que existe alguma comida brasileira com influência da culinária peruana? Se sim, qual ? '

# Lista das respostas para substituir por "Não sei"
respostas_incertas = ['não conheço a comida peruana para analisar sua influência no brasil',
                      'não sei dizer, mas deve ter, pela diversidade e variedades de milho.',
                      'acho que sim, mas não sei qual é.', 'não sei dizer na verdade', 'sei que tem',
                      'não lembro', 'não sei', 'nao sei dizer', 'acho que sim não sei qual.',
                      'não sei', 'nao sei', 'com certeza tem, mas nao sei informar qual.'                    
]

df[coluna] = df[coluna].str.lower().str.strip()
df[coluna] = df[coluna].replace(respostas_incertas, 'não sei')

inf_br = df[coluna].value_counts()

print("Respostas à pergunta sobre influência da culinária peruana na brasileira':\n")
print(f"Total de Respostas: {inf_br.sum()}")
print(f"Respostas 'Não': {inf_br['não']} ({inf_br['não']/inf_br.sum()*100:.1f}%)")
print(f"Respostas 'Não sei': {inf_br['não sei']} ({inf_br['não sei']/inf_br.sum()*100:.1f}%)")
print(f"Menções de 'Pamonha' como influência peruana: {inf_br['pamonha']} ({inf_br['pamonha']/inf_br.sum()*100:.1f}%)")

for comida, count in inf_br.items():
    if comida not in ['não', 'não sei', 'pamonha'] and count >= 1:
        print(f"Outras menções relevantes: '{comida}' com {count} ocorrências ({count/inf_br.sum()*100:.1f}%).")

plt.figure(figsize=(20, 12))
inf_br.plot(kind='bar')
plt.title('Respostas à pergunta sobre influência da culinária peruana na brasileira')
plt.xlabel('Respostas')
plt.ylabel('Número de ocorrências')
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('./Gráficos/Existe alguma comida brasileira com influência da culinária peruana.png')
plt.show()
```
